### Confusion Matrix:

**Explanation 1**

This table is used to identify, how your classification model is performing by compairing it's predicted value and actual value. It breaks down the predictions into four categories: correct predictions for both classes (true positives (TP) and true negatives (TN)) and incorrect predictions (false positives (FP) and false negatives (FN)). This helps you understand where the model is making mistakes, so you can improve it on the scale of positive (P) and negative (N).



We will more understand this by taking example of Corona virus. Now lab tested all 100 and 50 of them are positive and 50 of them are negative based on prediction and lab test. Now consider following cases:
1. By symptoms, some of them appeared to be positive (Predicted) and lab result also confirmed (Actual) that they are positive i.e. TP.
2. By symptoms, some of them appeared to be negative (Predicted) and lab result also confirmed (Actual) that they are negative i.e. TN
3. By symptoms, some of them actually positive (Predicted) but lab result confirmed (Actual) that they are negative i.e. FN
4. By symptoms, some of them are actually negative (Predicted) but lab result confirmed (Actual) that they are positive i.e. FP

By using above concept we will try to understand two classification model here:
1. Recall:

    $\frac{TP}{TP+FN} = \frac{TP}{Actual Positive}$ here let's say 50 were total tested as positive and among them 30 is correctly diagnosted and actually proven to be positive.

    So, 30 / 50 = 0.6

2. Precision

    $\frac{TP}{TP+FP} = \frac{TP}{Predicted Positive}$, here let's say 60 were predicted positive, and among how many was actually found positive.

    So, 60 / 50 = 1.2

```python
# Source code credit for this function: https://gist.github.com/shaypal5/94c53d765083101efc0240d776a23823
def print_confusion_matrix(confusion_matrix, class_names, figsize = (10,7), fontsize=14):
    """Prints a confusion matrix, as returned by sklearn.metrics.confusion_matrix, as a heatmap.
    
    Arguments
    ---------
    confusion_matrix: numpy.ndarray
        The numpy.ndarray object returned from a call to sklearn.metrics.confusion_matrix. 
        Similarly constructed ndarrays can also be used.
    class_names: list
        An ordered list of class names, in the order they index the given confusion matrix.
    figsize: tuple
        A 2-long tuple, the first value determining the horizontal size of the ouputted figure,
        the second determining the vertical size. Defaults to (10,7).
    fontsize: int
        Font size for axes labels. Defaults to 14.
        
    Returns
    -------
    matplotlib.figure.Figure
        The resulting confusion matrix figure
    """
    df_cm = pd.DataFrame(
        confusion_matrix, index=class_names, columns=class_names, 
    )
    fig = plt.figure(figsize=figsize)
    try:
        heatmap = sns.heatmap(df_cm, annot=True, fmt="d")
    except ValueError:
        raise ValueError("Confusion matrix values must be integers.")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    plt.ylabel('Truth')
    plt.xlabel('Prediction')
```

```python
truth =      ["Dog","Not a dog","Not a dog","Dog", "Dog", "Not a dog", "Not a dog", "Dog","Dog", "Not a dog"]
prediction = ["Dog","Dog","Not a dog","Dog","Dog", "Dog", "Not a dog", "Not a dog", "Dog", "Not a dog"]
```

```python
cm = confusion_matrix(truth,prediction)
print_confusion_matrix(cm,["Dog","Not a dog"])
```

This Linear regression can be categorized in two ways:

y=α<sub>0</sub> + α<sub>1</sub>x<sub>1</sub>

## K-Means Clustering

K-means is an iterative algorithm that assigns data points to clusters and updates centroids until convergence. In Clustering, we are identifying the groups for similar data points. Here K denotes number of clusters. We can perform some steps for this algorithm: 

Step 1: Find out **Centroids** for each clusters. 
* Random Selection: Pick k data points randomly from the dataset as initial centroids.
* K-Means++: A smarter initialization that spreads centroids out by choosing points with probability proportional to their distance from already selected centroids (default in scikit-learn). 

Step 2: Points that are nearest to the Centroid, we keep them in one group. \
Step 3: Move the Centroids by calculating the Mean 

### How do we select the value of K?

We can find out the value of K by using **WCSS** (Within cluster Sum of Squares). In this process we are initializing K from 1 to 20 (means 1 Centroid to 20 Centroid) and keep checking the distances** with each data point of the K using below formula:

$WCSS = \displaystyle\sum_{i=1}^K (\displaystyle{Distance between points to the nearest Centroid})^2$


** Finding distance we can use Eucledian or Manhattan method. Euclean disance said, if we have two points called P1 and p2 then distance would be

$ED = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$

We will notice when we increase K in each iteration, our WCSS value decreasing and we get graph like below:

### Random Initialization Trap

In some cases, your data points looks like similar but can become part of different cluster and not properly clustered because of Centroids are located very near to each other. To resolve this issue we can use **K-Means++** technique. In this technique we keep Centroid in far distance so that we can have clearly separate clusters without any overlapping.

## Hierarchical Clustering

In Hierarchical clustering there is no centroid like K-Means. To create cluster we can use two methods:

* Agglomerative
* Divisive

Steps for Agglomerative technique:

Step 1: For each point initially we will consider it as a separate cluster. \
Step 2: Find the nearest point and create a new cluster. \
Step 3: Keep repeat Step 2 until we get one cluster.


To find out how many exactly cluster we are getting by this technique, we can use **Dendogram** like we are seeing in above diagram.

## DBSCAN Clustering

In this diagram, minPts = 4. Point A and the other red points are core points, because the area surrounding these points in an ε radius contain at least 4 points (including the point itself). Because they are all reachable from one another, they form a single cluster. Points B and C are not core points, but are reachable from A (via other core points) and thus belong to the cluster as well. Point N is a noise point that is neither a core point nor directly-reachable.

In other words, if we select randomly a point and draw a circle using Radius (ϵ), 

**Core points** if we have atleast 4 (which is minPts) within circle, then that point called core points. \
**Border Point** is when no. of data points within the radius will be less than minPts. \
**Outlier Point** When there is no other points within radius.

## Silhouette Clustering

This is used measure performance in unsupervised learning always range between -1 to 1. When you create one cluster and many data points, then we compute a(i) from every cluster. We simply take any one data point and calculate distance of each datapoint and find mean after sum of all.

Then from other cluster b(i) calculate distance of each data point to first cluster datapoint and calculate the data point from first cluster point to second cluster point and find avg.

# K Means Clustering Algorithms Implementation

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=23)
```

```python
X.shape
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
#Elbow method to select i value **MANUALLY**
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# elbow method
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++')
y_labels = kmeans.fit_predict(X_train)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1],c=y_labels) #train data category
```

```python
y_test_labels = kmeans.predict(X_test)
```

```python
plt.scatter(X_test[:, 0], X_test[:, 1],c=y_test_labels) #test data category
```

```python
## Peformance matrix - silhouette score
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(X_train)
    score = silhouette_score(X_train, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.title('Silhouette Coefficient')
plt.xlabel('Number of clusters')   
plt.ylabel('Silhouette Coefficient')
plt.show()
```

as you can see, the for cluster 3, it's more the 80% coff. value i.e. more closer to +1 have better performance

# Hierarchical Clustering

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
iris_data
```

We can see here we have four features which needs to convert them to 2D (by using PCA) and needed to apply **feature scaling**.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(iris_data)
```

```python
x_scaled
```

```python
x_scaled.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_scaled = pca.fit_transform(x_scaled)
```

```python
pca_scaled
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c=iris.target)
```
